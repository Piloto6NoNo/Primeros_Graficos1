```python
#Importacion de librerias
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw_google_logo(): #Se define la función#

    fig, ax = plt.subplots(figsize=(9, 2)) #Se crea un o el area de dibujo
    ax.set_facecolor("white") #sera el fondo
    ax.axis("off") #oculta los ejes#

#A continuacion se define un dicionario de colores, cada letra del logo
    colors = {
        "G": "blue",
        "o1": "red",
        "o2": "yellow",
        "g": "blue",
        "l": "green",
        "e": "red",
        "!": "blue"}

    ax.text(0.05, 0.5, "G", fontsize=120, color=colors["G"], weight="bold", va="center")  #0.05, 0.5: son las coordenadas x y y (dónde poner la letra dentro del gráfico).
    ax.text(0.25, 0.5, "o", fontsize=120, color=colors["o1"], weight="bold", va="center") #"G": es el texto que se va a dibujar.
    ax.text(0.42, 0.5, "o", fontsize=120, color=colors["o2"], weight="bold", va="center") #fontsize=120: tamaño de la letra (muy grande).
    ax.text(0.59, 0.5, "g", fontsize=120, color=colors["g"], weight="bold", va="center") #color=colors["G"]: color azul (según el diccionario).
    ax.text(0.75, 0.5, "l", fontsize=120, color=colors["l"], weight="bold", va="center") #weight="bold": la letra es negrita.
    ax.text(0.82, 0.5, "e", fontsize=120, color=colors["e"], weight="bold", va="center") #va="center": alinea verticalmente la letra en el centro.
    ax.text(0.95, 0.5, "!", fontsize=120, color=colors["!"], weight="bold", va="center")
    plt.show() #abre una ventana con la figura que contiene el logo
# Ejecuta todo lo que definio en la parte de arriba
draw_google_logo()
```
